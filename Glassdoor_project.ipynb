## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df.head()
````

````python
#replacing -1 with Nan to identify the NUll Values, inplace is used to modify existing
df.replace('-1',np.nan,inplace=True)
#counting the number of null values
df.isnull().sum().sum()
````

````python
#Replacing Unknown values in columns with Null value
df["Revenue"] = df["Revenue"].replace(["Unknown / Non-Applicable"], np.nan)
df["Type of ownership"] = df["Type of ownership"].replace(["Unknown"], np.nan)
df["Size"] = df["Size"].replace(["Unknown"], np.nan)
````

````python
# Removing missing values and filtering out invalid salaries
df = df[df["Salary Estimate"].notna() & (df["Salary Estimate"] != "-1")]

````

````python
# Cleaning Salary Estimate column
df["Salary Estimate"] = df["Salary Estimate"].str.split("(").str[0]  # Remove text in parentheses
df["Salary Estimate"] = df["Salary Estimate"].str.replace("K", "").str.replace("$", "").str.replace("Per Hour", "").str.replace("Employer Provided Salary:", "").str.strip()
````

````python
# Handling salary ranges
def extract_salary(salary_str):
    parts = salary_str.split("-")
    try:
        min_salary = int(parts[0].strip()) if len(parts) > 1 else int(parts[0].strip())
        max_salary = int(parts[1].strip()) if len(parts) > 1 else min_salary
        return (min_salary, max_salary)
    except:
        return (None, None)

df["Min Salary"], df["Max Salary"] = zip(*df["Salary Estimate"].apply(extract_salary))
df = df.dropna(subset=["Min Salary", "Max Salary"])  # Removing invalid rows
df["Avg Salary"] = (df["Min Salary"] + df["Max Salary"]) / 2  # Calculating average salary
````

````python

# Grouping by Job Title and calculating the mean salary
top_jobs = df["Job Title"].value_counts().head(10).index  # Selecting top 10 job titles
avg_salary_per_title = df[df["Job Title"].isin(top_jobs)].groupby("Job Title")["Avg Salary"].mean()
````

````python
# Grouping by Location and calculating the mean salary
top_locations = df["Location"].value_counts().head(10).index  # Selecting top 10 locations
avg_salary_per_location = df[df["Location"].isin(top_locations)].groupby("Location")["Avg Salary"].mean()
````
